# Line graph using matplotlib

```python
import matplotlib.pyplot as mp
```

```python
mp.plot(data['Sale Price'])
```

```python
mp.plot(data['Sale Price'], color = 'orange' )
mp.xlabel('record id')
mp.ylabel('selling price')
mp.title('sale price line graph')
mp.show()
```

# Pie chart using mp (can be used for categorical vs numerical variable)

```python
data.info()
```

```python
data.groupby('Condition of the House')['ID'].count()
```

```python
val=(30, 1701, 14031, 5679, 172)
lab=('Bad', 'Excellent', 'Fair', 'Good', 'Okay')
```

```python
mp.pie(val, labels= lab)
mp.title('Condition of the House VS no of houses')
mp.show()
```

# Bar graph using mp (used for categorical vs numerical variables) 

```python
mp.bar(lab,val,color='orange')
mp.xlabel('Condition of the House')
mp.ylabel('no of houses')
mp.title('Condition of the House VS no of houses')
mp.show()
```

# Scatter plot using mp (used to know relation between two numeric variables)

```python
data.columns
```

```python
mp.scatter(x=data['Flat Area (in Sqft)'], y=data['Sale Price'], color='orange')
mp.xlabel('Flat Area (in Sqft)')
mp.ylabel('Sale Price')
mp.title('Sale Price vs Area')
mp.show()
```

# Histogram(used to plot freq over a range of values)

# Can be used to check whether the data is normally distributed ot not

# Data should be divides into bins

```python
data.describe()
```

```python
mp.hist(data['Flat Area (in Sqft)'], bins =50)
```

```python
mp.hist(data['Age of House (in Years)'], bins=20, color='orange')
mp.xlabel('Age of House (in Years)')
mp.ylabel('no of houses')
mp.title('frequency distribution')
mp.show()
```
