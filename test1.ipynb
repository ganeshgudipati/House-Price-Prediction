# treating outliers

```python
data['Sale Price'].describe()
```

```python
plot.scatter(x=data['ID'],y=data['Sale Price'], color='blue')
```

```python
q1=data['Sale Price'].quantile(0.25)
q3=data['Sale Price'].quantile(0.75)
```

```python
IQR=q3-q1
IQR
```

```python
upper_limit= q3 + IQR*1.5
lower_limit= q1 - IQR*1.5
```

```python
upper_limit,lower_limit
```

```python
def impute(value):
    if value > upper_limit:
        return upper_limit
    if value < lower_limit:
        return lower_limit
    else:
        return value
```

```python
data['Sale Price']=data["Sale Price"].apply(impute)
```

```python
data['Sale Price'].describe()
```

# ANOVA

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
data=data.rename(columns={'Sale Price' : 'Sale_Price'})
```

```python
data=data.rename(columns={'Condition of the House' : 'Condition_of_the_House','Waterfront View' : 'Waterfront_View', 
                          'Ever Renovated':'Ever_Renovated' })
```

```python
data.head()
```

```python
data1=data
```

```python
data1
```

```python
mod=ols('Sale_Price ~ Condition_of_the_House ', data=data1).fit()
```

```python
anova_table=sm.stats.anova_lm(mod, typ=2)
print(anova_table)
```

```python
mod=ols('Sale_Price ~ Waterfront_View ', data=data1).fit()
```

```python
print(sm.stats.anova_lm(mod, typ=2))
```

```python
mod=ols('Sale_Price ~ Ever_Renovated ', data=data1).fit()
```

```python
print(sm.stats.anova_lm(mod, typ=2))
```

```python
mod=ols('Sale_Price ~ Zipcode ', data=data1).fit()
```

```python
print(sm.stats.anova_lm(mod, typ=2))
```
